# Задание 2 -Линейная аппроксимация методом наименьших квадратов
y = w * x + b

y : Payment

x : Amount of candy

w,b : Coefficients

```
x = df['no_candy'].values
y = df['payment'].values
```

```
plt.scatter(x,y)
plt.title('Зависимость y(x)')
plt.xlabel('Candy')
plt.ylabel('Payment')
plt.show()
```

Определить функцию ошибки

```
# функция для поиска средней квадратичной ошибки
def MSE(x,y,w,b):
  total_error = 0
  for i in range(len(x)):
    error = (y[i]-w*x[i]-b)**2/2
    total_error += error
  return total_error/len(x)
```

Обновить коэффициенты для минимизации функции ошибки методом градиентный спуск

```
# Фунция для поиска правильных коэффициентов
def fit(x,y,alpha):
  index = 0 # Число итераций
  # Создать случайный коэффициенты
  w0 = 1   
  b0 = 1
  while True:
    index +=1 
    rmse = math.sqrt(MSE(x,y,w0,b0))
    # Выводить корень из 2 от средней квадратичной ошибки
    print('Loop',index,': RMSE =',end=' ')
    print(rmse)
    # Найти градиент от функцию ошибки на коэффициентов
    w_gradient = 0
    b_gradient = 0
    i = 0
    for i in range(len(x)):
      w_gradient = w_gradient + (w0*x[i]+b0-y[i])*x[i]
      b_gradient = b_gradient + (w0*x[i]+b0-y[i])
    w_temp = w0 
    b_temp = b0
    # Обновить коэффициенты
    w0 -=  alpha*w_gradient/20  
    b0 -= alpha*b_gradient/20  
    if abs(w_temp-w0)<0.00001 or index>50 : # количество итераций не выше 50 и изменение w выше чем 0.00001
      break
  return w0,b0
```

```
w,b = fit(x,y,0.00005)
```

```
print("w=",w)
print("b=",b)
print("Результат : y = %5f * x + %5f "%(w,b))
```
