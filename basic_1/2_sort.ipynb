```
import numpy as np
import pandas as pd
from timeit import default_timer as timer
from datetime import timedelta
import datetime
```

```
unsorted_array = np.random.randint(0,2000,(1000)).astype(float)
```

```
data = unsorted_array.copy()
start = timer()
#	Сортировка выбором
for i in range(0,len(data)):
  index_min = i
  for j in range(index_min,len(data)):
    if data[index_min]>data[j]:
      data[index_min],data[j] = data[j],data[index_min]
end = timer()
t1 = datetime.timedelta(seconds=end-start).total_seconds()
print(t1)
```

```
data = np.insert(data,0,values=t1)
```

```
df1 = pd.DataFrame(data)
df1
```

```
data = unsorted_array.copy()
start = timer()
# Сортировка вставками
for i in range(1, len(data)): 
  key = data[i] 
  j = i-1
  while j >=0 and key < data[j] :
      data[j+1] = data[j] 
      j -= 1 
  data[j+1] = key 
end = timer()
t2 = datetime.timedelta(seconds=end-start).total_seconds()
print(t2)
```

```
data = np.insert(data,0,values=t2)
df2 = pd.DataFrame(data)
```

```
data = unsorted_array.copy()
# Сортировка “Методом пузырька”
start = timer()
n = len(data)
for i in range(n-1):  
  for j in range(0, n-i-1): 
    if data[j] > data[j+1] : 
      data[j], data[j+1] = data[j+1], data[j]
end = timer()
t3 = datetime.timedelta(seconds=end-start).total_seconds()
print(t3)
```

```
data = np.insert(data,0,values=t3)
df3 = pd.DataFrame(data)
```

```
#	Сортировка Шелла
data = unsorted_array.copy()
start = timer()
n = len(data) 
gap = n//2
while gap > 0: 
  for i in range(gap,n): 
    temp = data[i] 
    j = i 
    while  j >= gap and data[j-gap] >temp: 
      data[j] = data[j-gap] 
      j -= gap 
    data[j] = temp 
  gap //= 2
end = timer()
t4 = datetime.timedelta(seconds=end-start).total_seconds()
print(t4)
```

```
data = np.insert(data,0,values=t4)
df4 = pd.DataFrame(data)
```

```
# Быстрая сортировка
data = unsorted_array.copy()
def partition(data,low,high): 
    i = ( low-1 )         # индекс найменьшего элемента
    pivot = data[high]     # pivot 
  
    for j in range(low , high): 
  
        # Если текущий элемент меньше pivot 
        if   data[j] < pivot: 
          
            # увеличивать индекс у меньше элемента 
            i = i+1 
            data[i],data[j] = data[j],data[i] 
  
    data[i+1],data[high] = data[high],data[i+1] 
    return ( i+1 ) 
  
# data[] --> Массив должен сортирован
# low  --> начальный индекс, 
# high  --> конечный индекс
  
# функция для быстрой сортировки
def quickSort(data,low,high): 
    if low < high: 
  
        # pi is partitioning index, arr[p] is now 
        # at right place 
        pi = partition(data,low,high) 
  
        # Отдельно сортировать элементы перед разделом и после раздела
        quickSort(data, low, pi-1) 
        quickSort(data, pi+1, high) 
start = timer()
quickSort(data,0,len(data)-1)
end = timer()
t5 = datetime.timedelta(seconds=end-start).total_seconds()
print(t5)
```

```
data = np.insert(data,0,values=t5)
df5 = pd.DataFrame(data)
```

```
unsorted_array = np.insert(unsorted_array,0,np.NaN)
df = pd.DataFrame(unsorted_array)
```

```
df_ = pd.concat([df,df1,df2,df3,df4,df5],axis=1)
```

```
excel_file = df_.to_excel('sorted_array.xlsx',header=["Начальный массив",
                                                      "Сортировка выбором",
                                                      "Сортировка вставками",
                                                      "Сортировка “Методом пузырька",
                                                      "Сортировка Шелла",
                                                      "Быстрая сортировка"]
                          ,columns=None,index=False)
```
