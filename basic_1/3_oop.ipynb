# Задание 1 : 
Создать класс ThreeStates и определить у него один атрибут state и метод
hasstate. Атрибут реализовать как приватный. В рамках метода реализовать
переключение состояний: первое, второе, третье. Продолжительность первого
состояния составляет 5 секунд, второго — 3 секунды, третьего — 2 секунды.
Переключение между состояниями должно осуществляться только в
указанном порядке (первое, второе, третье). Проверить работу примера, создав
экземпляр и вызвав описанный метод. 


```
import time
from timeit import default_timer as timer
from datetime import timedelta
import datetime
```

```
class ThreeStates:
  def __init__(self,state='Первое состояние'):
    self.__privateState = state
    self.start = timer()
  def current_State(self):
    end = timer()
    t = float(datetime.timedelta(seconds=end-self.start).total_seconds())
    print(t,'seconds :',end=' ')
    if t%10 <= 5:
      self.__privateState='Первое состояние'
    elif t%10>5 and t%10<=8:
      self.__privateState='Второе состояние'
    else:
      self.__privateState = 'Третье состояние'
    return self.__privateState
```

```
s = ThreeStates()
def print_State(s):
  c = 0
  while True:
    print(s.current_State())
    time.sleep(1)
    c = c+1
    if c>20:
      break
```

```
print_State(s)
```

# Задание 2
Реализовать класс Volume, в котором определить атрибуты: length (длина),width (ширина) и height (высота). Значения данных атрибутов должны передаваться при  создании экземпляра класса. Атрибуты сделат защищенными. Определить метод расчета, использовав формулу: длина* ширина * высота. Проверить работу метода.

```
class Volume:
  def __init__(self,length=1,width=1,height=1):
    self._length = length
    self._width  = width
    self._height = height
  def volume(self):
    return self._length*self._width*self._height
```

```
v = Volume(2,3,5.5)
print(v.volume())
```

# Задание 3
Реализовать базовый класс Employee, в котором определить атрибуты: name,
patronymic, surname, salary. Последний атрибут должен быть защищенным и
ссылаться на словарь, содержащий элементы: жалование и бонус, например,
{"wage": wage, "bonus": bonus}. Создать класс Salary на базе класса Employee.
В классе Salary реализовать методы получения полного имени сотрудника
(get_full_name) и дохода с учетом премии (get_total_income). Создать
экземпляры класса Salary, передать данные, проверить значения атрибутов,
вызвать методы экземпляров.

```
class Employee:
  def __init__(self,name,patronymic,surname,salary):
    self.name = name
    self.patronymic = patronymic
    self.surname = surname
    self._salary = salary
```

```
class Salary(Employee):
  def __init__(self,name,patronymic,surname,salary):
    Employee.__init__(self,name,patronymic,surname,salary)
  def get_full_name(self):
    return self.surname + " " + self.name + " " + self.patronymic
  def get_total_income(self):
    return self._salary["wage"]+self._salary["bonus"]
```

```
s = Salary("Than","Quoc","Dat",{"wage":1000,"bonus":70})
print(s.get_full_name())
print(s.get_total_income())
```

# Задание 4
 Реализуйте базовый класс Airplane. У данного класса должны быть
следующие атрибуты: speed, color, name, is_jet (булево). А также методы: go,
stop, direction, которые должны сообщать, что самолет летит, не летит,
повернул в полете. Опишите несколько дочерних классов самолетов:
FastAirplane, Biplane, ArmyAirplane. Добавьте в базовый класс метод
show_speed, который должен показывать текущую скорость самолета. Для
класса FastAirplane переопределите метод show_speed. При значении скорости
свыше 1300 (FastAirplane) должно выводиться сообщение о сверхзвуковой
скорости. Создайте экземпляры классов, передайте значения атрибутов.
Выполните доступ к атрибутам, выведите результат. Выполните вызов
методов и также покажите результат

```
class Airplane:
  def __init__(self,speed,color,name,is_jet=True):
    self.speed = speed
    self.color = color
    self.name = name
    self.is_jet = is_jet
  def show_speed(self):
    print(self.speed)
  def go(self):
    if self.speed != 0:
      return True
    else :
      return False
  def stop(self):
    if self.speed == 0:
      return True
    else :
      return False
  def direction(self):
    if self.speed >0:
      print("Не вернул")
    elif self.speed<0:
      print("Повернул")
    else :
      print("Стоп")
```

```
class FastAirplane(Airplane):
  def __init__(self,speed,color,name,is_jet=True):
    Airplane.__init__(self,speed,color,name,is_jet=True)
  def show_speed(self):
    if abs(self.speed) > 1300 :
      print("Скорость самолета было более 1300 км / ч.Это сверхзвуковая скорость!")
    else :
      Airplane.show_speed(self)
```

```
class BiAirplane(Airplane):
  def __init__(self,speed,color,name,is_jet=False):
    Airplane.__init__(self,speed,color,name,is_jet=False)
class ArmyPlane(Airplane):
  def __init__(self,speed,color,name,is_jet=True):
    Airplane.__init__(self,speed,color,name,is_jet=True)
```

```
# FastAirplane
FA = FastAirplane(-1400,"White","DAT99")
FA.show_speed()
print(FA.go())
print(FA.stop())
FA.direction()
```

```

#BiAirPlane
BA = BiAirplane(700,"Red","DAT15")
BA.show_speed()
print("Go :",BA.go())
print("Stop :",BA.stop())
BA.direction()
```

```
AA = ArmyPlane(0,"Red","DAT15")
AA.show_speed()
print("Go :",AA.go())
print("Stop :",AA.stop())
AA.direction()
```

Создать три дочерних класса my_sum (сложение), my_sub (вычитание),

метода calc. Для каждого из классов метод должен выполнять математическую
операцию, соответствующую названию класса. Создать экземпляры классов и
проверить работу методов.

```
#MathOperations Class
class MathOperations:
  def __init__(self,first,second):
    self.first_num = first
    self.second_num = second
  def calc(self):
    print("Запуск операции")
    pass
# my_sum class  
class my_sum(MathOperations):
  def __init__(self,first,second):
    MathOperations.__init__(self,first,second)
  def calc(self):
    return self.first_num+self.second_num
# my_sub class
class my_sub(MathOperations):
  def __init__(self,first,second):
    MathOperations.__init__(self,first,second)
  def calc(self):
    return self.first_num-self.second_num
# my_mult class
class my_mult(MathOperations):
  def __init__(self,first,second):
    MathOperations.__init__(self,first,second)
  def calc(self):
    return self.first_num*self.second_num
```
