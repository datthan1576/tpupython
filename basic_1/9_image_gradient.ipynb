```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(r"/content/drive/MyDrive/Colab/Метод машинного обучения/meme.jpg")
plt.imshow(img)
```

```python
import numpy as np

img_gray = np.mean(img, axis=-1)
plt.imshow(img_gray, cmap='gray')
```

```python
sobel_matrix_1 = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])/4
sobel_matrix_2 = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])/4

sob_image = np.zeros(img_gray.shape)
for i in range(1, len(sob_image)-1):
  for j in range(1, len(sob_image[0])-1):
    sob_image[i, j] = (np.abs(np.sum(img_gray[i-1:i+2, j-1:j+2] * sobel_matrix_1)) +
                       np.abs(np.sum(img_gray[i-1:i+2, j-1:j+2] * sobel_matrix_2))) / 2

plt.imshow(sob_image, cmap='gray')
```
